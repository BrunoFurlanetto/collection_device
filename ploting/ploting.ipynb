```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Define empty list for axes and locations of files to be read

```python
visual_file_test = '../choice.dat'
auditory_file_test = '../simple.dat'
tactile_file_test = '../tactile_test.dat'

x = []
y = []
```

#### Reading the file and defining the values ​​of the axes

```python
with open(auditory_file_test, 'r') as f:
    lines = f.readlines()

for line in lines:
    data = line.strip().split()
    x.append(data[0])
    y.append(float(data[1]))
```

#### Calculate average of values

```python
mean = np.mean(y)
mean_line = [mean] * len(y)
```

#### Plot graph with error bars and mean line and set axis labels

```python
plt.bar(x, y, yerr=np.std(y))
plt.plot(x, mean_line, linestyle='-', color='red', label=f'Média: {mean:.2f}')
plt.text(7, mean + 0.1, f'Mean: {mean:.2f}', color='red', ha='center', va='bottom')
plt.title('Visual reaction time')
plt.xlabel('Attempt')
plt.ylabel('Reaction time (s)')
plt.legend(['Mean'])
plt.show()
```
